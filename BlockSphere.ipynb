```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_qsphere, plot_bloch_multivector
from qiskit.quantum_info import Statevector
```

# Plot $|0\rangle$

```python
qc = QuantumCircuit(1)
Statevector(qc)
```

```python
plot_bloch_multivector(qc)
```

# Plot $|1\rangle$

```python
qc = QuantumCircuit(1)
qc.x(0)
Statevector(qc)
```

```python
plot_bloch_multivector(qc)
```

# Plot $\frac{1}{\sqrt{2}} (|0\rangle + |1\rangle)$

```python
qc = QuantumCircuit(1)
qc.h(0)
Statevector(qc)
```

```python
plot_bloch_multivector(qc)
```

# Plot $\frac{1}{\sqrt{2}} (|0\rangle + e^{i \frac{\pi}{4}} |1\rangle)$

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.p(np.pi/4, 0)
Statevector(qc)
```

```python
plot_bloch_multivector(qc)
```

# Plot $\frac{i}{\sqrt{2}} (-|0\rangle + |1\rangle$

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.y(0)
Statevector(qc)
```

```python
plot_bloch_multivector(qc)
```

# Plot $\frac{1}{\sqrt{2}} (|0\rangle - |1\rangle)$

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.z(0)
Statevector(qc)
```

```python
plot_bloch_multivector(qc)
```

# Perform an Euler Rotation

```python
theta = np.pi/3
phi = 3*np.pi/2
lamb = 0
```

```python
qc = QuantumCircuit(1)
qc.u(theta, phi, lamb, 0)
Statevector(qc)
```

```python
plot_bloch_multivector(qc)
```
